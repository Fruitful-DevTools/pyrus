```python
from bs4 import BeautifulSoup
import requests

from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:80.0) Gecko/20100101 Firefox/80.0'}

page = requests.get('https://www.linkedin.com/in/seancharlton/recent-activity/all/', headers=headers, stream=True)

soup = BeautifulSoup(page.content, 'html5')

print(soup)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Set up the Selenium Chrome driver with options to run headless
options = Options()
options.headless = True
driver = webdriver.Chrome(options=options)

# Make a request to a website
driver.get('https://www.linkedin.com/in/seancharlton/recent-activity/all/')

# Get the page source
page_source = driver.page_source

# Print the page source
print(page_source)
```

```python
import requests
from bs4 import BeautifulSoup

# Set the URL of the LinkedIn profile
url = 'https://www.linkedin.com/in/seancharlton/'

# Make a request to the URL
response = requests.get(url)

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the text of the LinkedIn profile's headline
headline = soup.find('h2', {'class': 'mt1'}).text.strip()

# Print the headline
print(headline)
```
